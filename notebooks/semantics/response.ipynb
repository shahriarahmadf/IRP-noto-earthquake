```python
import requests

def get_location_semantics(lat, lon):
    url = "https://nominatim.openstreetmap.org/reverse"
    params = {
        'lat': lat,
        'lon': lon,
        'format': 'json',
        'addressdetails': 1
    }

    headers = {
        'User-Agent': 'LocationSemanticsApp/1.0 (your_email@example.com)'  # Replace with your email
    }
    try:
        response = requests.get(url, params=params, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
        output = set([data['class'], data['type'], data['addresstype']])
        output = [item for item in output if item != 'unclassified']
        result = ' '.join(output) 
        print('returning')
        return result
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
    except ValueError:
        print("Failed to parse JSON response")
    return None

# latitude = 37.012113
# longitude = 136.780784

# location_data = get_location_semantics(latitude, longitude)
# print(type(location_data))  # This should print <class 'dict'> if the request is successful and returns JSON data
# print(location_data)  # Print the actual data or error message

```
